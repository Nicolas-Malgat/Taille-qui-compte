## Visualisation des effets de la data augmentation

```python
augmented_datagen = ImageDataGenerator(
 rotation_range=10,
 width_shift_range=0.1,
 height_shift_range=0.1,
 shear_range=0.2,
 zoom_range=0.2,
 horizontal_flip=True,
 brightness_range=(0.3, 1.25),
 fill_mode='nearest')

train_car_dir = "..\datas\CURATED\\taillequicompte\\train_airplane_automobile_bird_cat_deer_dog_frog_horse_ship_truck\\automobile"
```
